```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('C:/Users/tangt/Desktop/Movie-Prediction-Rottentomatoes/movie.csv')
df.head()
```

```python
correlation = df.corr()
plt.figure(figsize=(25,25))
colormap = plt.cm.RdBu
h = sns.heatmap(correlation,linewidths=0.1,vmax=1,vmin=-1,square=True,annot=True,cmap=colormap,linecolor='white',xticklabels=True,yticklabels=True)
plt.title('Correlation between different factors')
```

```python
plt.figure(figsize=(20,20))
plt.rcParams.update({'font.size': 15})
c_score=df['critic_score'].str.strip("%").astype(float)
plt.scatter(x=df['theater_year'], y=c_score,alpha=0.4)
plt.xlabel('theater_year')
plt.ylabel('critic_score')
plt.xticks(np.arange(1920, 2020, step=3), rotation=30)
plt.axvline(x=1990, linestyle='dashed')
```
